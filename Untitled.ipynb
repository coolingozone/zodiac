```python
import numpy as np
import pandas as pd

df=pd.read_csv('The Zodiac Race Challenge_Dataset.csv')
df.head()
```

```python
df.drop(24246, inplace=True)
```

```python
df = df.astype({"Desert": float, "Mountain": float,"Water":float,"Land":float,"Forest":float,"Snow":float})

df.info()
```

```python
from sklearn import preprocessing 
le = preprocessing.LabelEncoder() #create an encoder instance
le.fit(["No", "Yes"])             #fit the encoder with two string "no" and "yes"
print(list(list(le.classes_)))
df['Is_Raining']=le.transform(df['Is_Raining'])    #to transform df['driveway'] value of yes and no to 1 and 0
df['Had_A_Good_Sleep']=le.transform(df['Had_A_Good_Sleep'])    #to transform df['driveway'] value of yes and no to 1 and 0
print(df.head())
```

```python
le2 = preprocessing.LabelEncoder() #create an encoder instance
le2.fit(['rat','ox','tiger','rabbit','monkey','rooster','dog','pig','dragon','snake','horse','goat'])             #fit the encoder with two string "no" and "yes"
print(list(list(le2.classes_)))
df['Winner']=le2.transform(df['Winner'])    #to transform df['driveway'] value of yes and no to 1 and 0
df['Second']=le2.transform(df['Second'])    #to transform df['driveway'] value of yes and no to 1 and 0
df['Third']=le2.transform(df['Third'])    #to transform df['driveway'] value of yes and no to 1 and 0
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
import  seaborn as sns
import matplotlib.pyplot as plt

Y = df['Winner']  #price is the target, Y
X=np.array([df['Desert'], df['Mountain'],df['Water'],df['Land'],df['Forest'],df['Snow'],df['Average_Temperature'],df['Is_Raining'],df['Had_A_Good_Sleep']])
X=X.T
for row in X:
    
    total=row[0]+row[1]+row[2]+row[3]+row[4]+row[5]
    row[0]=row[0]/total
    row[1]=row[1]/total
    row[2]=row[2]/total
    row[3]=row[3]/total
    row[4]=row[4]/total
    row[5]=row[5]/total
big=max(X[:,6])
X[:,6]=X[:,6]/big

classifier=SGDClassifier(max_iter=2000)
Xtrain, Xtest, ytrain, ytest=train_test_split(X,Y,random_state=10,test_size=0.2)
classifier.fit(Xtrain,ytrain)
y_pred=classifier.predict(Xtest)
print(accuracy_score(ytest,y_pred))  #percentage of classification on the test data is correct   
mat2=confusion_matrix(ytest,y_pred)
sns.heatmap(mat2,square=True,annot=True,cbar=False,fmt='d')
plt.xlabel('predicted value')
plt.ylabel('true value')
plt.show()
```

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import confusion_matrix


clf = LinearSVC(random_state=20, tol=1e-5)
clf.fit(Xtrain, ytrain)
y_pred=clf.predict(Xtest)

print(accuracy_score(ytest,y_pred))


mat2=confusion_matrix(ytest,y_pred)
sns.heatmap(mat2,square=True,annot=True,cbar=False,fmt='d')
plt.xlabel('predicted value')
plt.ylabel('true value')
plt.show()
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.layers import Conv2D, MaxPooling2D
from keras.utils import np_utils
from keras.datasets import mnist
from sklearn.metrics import confusion_matrix
```

```python

Xtrain, Xtest, ytrain, ytest=train_test_split(X,Y,random_state=10,test_size=0.2)
#  Preprocess class labels
Y_train = np_utils.to_categorical(ytrain, 12)
Y_test = np_utils.to_categorical(ytest, 12)
print(Y_train.shape)

#  Define model architecture

model = Sequential()
model.add(Dense(512, input_shape=(9,)))
model.add(Activation('relu')) # An "activation" is just a non-linear function applied to the output
                              # of the layer above. Here, with a "rectified linear unit",
                              # we clamp all values below 0 to 0.
                           
model.add(Dropout(0.2))   # Dropout helps protect the model from memorizing or "overfitting" the training data
model.add(Dense(1024,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(1024,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(512,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(12,activation='softmax'))# This special "softmax" activation among other things,
                                 # ensures the output is a valid probaility distribution, that is
                                 # that its values are all non-negative and sum to 1.
# Compile model
model.compile(loss='categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
 
# Fit model on training data

history=model.fit(Xtrain, Y_train,
          batch_size=64, epochs=150)
 
# Evaluate model on test data
score = model.evaluate(Xtest, Y_test, verbose=0)
predictions = model.predict(Xtest)
confusion  = confusion_matrix(Y_test.argmax(axis=1), predictions.argmax(axis=1))
plt.figure(figsize = (5,5))
sns.heatmap(confusion,square=False,annot=True,cbar=False,fmt='d')
plt.xlabel('predicted value')
plt.ylabel('true value')
plt.show()
```
